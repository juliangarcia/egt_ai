```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
ir_data = pd.read_csv('reciprocity_data.csv', index_col=0)
```

```python
ir_data.head()
```

```python
import seaborn
```

```python
seaborn.set_context("talk", font_scale=0.8)
seaborn.set_style("whitegrid")
plt.plot(ir_data.index, ir_data["SJ"], label="Fixed social norm") # linestyle='dashed', color="black",
plt.plot(ir_data.index, ir_data["BOTTOMUP"], label="Endogenous social norms") # linestyle='solid',color="black",
plt.legend(loc="center right")
plt.xlabel("Population size")
plt.ylabel("Percentage of cooperation");
plt.savefig("figure4_irv2.pdf", bbox_inches="tight")
```
